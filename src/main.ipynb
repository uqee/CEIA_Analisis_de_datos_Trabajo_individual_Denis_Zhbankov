```python
import geodatasets
import geopandas
import geopandas.datasets
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn
import urllib.request
```

# 1. Exploración y comprensión de los datos

## 1.1. Descargar

```python
filename = 'df.csv'
if not os.path.exists(filename):
  urllib.request.urlretrieve('https://data.nasa.gov/docs/legacy/meteorite_landings/Meteorite_Landings.csv', filename)
df = pd.read_csv(filename)
```

## 1.2. Verificar

```python
df.head()
```

```python
df.tail()
```

## 1.3. Optimizar

```python
df.info()
```

```python
df.drop(columns=['GeoLocation'], inplace=True)
df.dropna(inplace=True)
df = df[sorted(df.columns)]
df = df.astype({
    'fall': 'category',
    'id': 'int32',
    'mass (g)': 'float32',
    'nametype': 'category',
    'recclass': 'category',
    'reclat': 'float32',
    'reclong': 'float32',
    'year': 'int32',
})
df.info()
```

# 2. Estadística descriptiva

```python
df.describe()
```

## 2.1. Variables numéricas

```python
_, ax = plt.subplots(1, 2, figsize=(16, 9))

ys = df['year']
ax[0].axvline(ys.mean(), color='green', ls='--', lw=2, label='Media')
ax[0].axvline(ys.median(), color='yellow', ls='--', lw=2, label='Mediana')
ax[0].axvline(ys.mode()[0], color='red', ls='--', lw=2, label='Moda')
ax[0].hist(ys, alpha=0.5, bins=100)
ax[0].legend()
ax[0].set_title('Años')
ax[0].set_xlabel('Año del impacto')
ax[0].set_ylabel('Número de impactos')
ax[0].set_yscale('log')

ms = df['mass (g)']
ax[1].axvline(ms.mean(), color='green', ls='--', lw=2, label='Media')
ax[1].axvline(ms.median(), color='yellow', ls='--', lw=2, label='Mediana')
ax[1].axvline(ms.mode()[0], color='red', ls='--', lw=2, label='Moda')
ax[1].hist(ms, alpha=0.5, bins=100)
ax[1].legend()
ax[1].set_title('Masas')
ax[1].set_xlabel('Masa del meteorito (gramos)')
ax[1].set_xscale('log')
ax[1].set_ylabel('Número de impactos')
ax[1].set_yscale('log')

plt.show()
```

```python
_, ax = plt.subplots(figsize=(16, 9))

ax.set_axis_off()
ax.set_title('Mapa de Impactos: Color (Año), Tamaño (Log Masa)')

geoDataFrame: geopandas.GeoDataFrame = geopandas.read_file(geodatasets.get_path('naturalearth.land'))
geoDataFrame.plot(ax=ax, color='#e0e0e0', edgecolor='#ffffff')

pathCollection = ax.scatter(df['reclong'], df['reclat'], alpha=0.75, c=df['year'], cmap='coolwarm', edgecolors='none', s=np.log(1 + df['mass (g)']))
plt.colorbar(pathCollection, shrink=0.5)

plt.show()
```

## 2.2. Variables categóricas

```python
df.describe(include='category')
```

```python
_, ax = plt.subplots(1, 3, figsize=(16, 9))

for i, column in enumerate(['fall', 'nametype', 'recclass']):
    counts = df[column].value_counts(normalize=True)

    mask = counts < 0.01
    if mask.any():
        counts_filtered = counts[~mask].copy()
        counts_filtered['Otros'] = counts[mask].sum()
    else:
        counts_filtered = counts

    ax[i].pie(counts_filtered, autopct='%1.1f%%', labels=counts_filtered.index, startangle=90)
    ax[i].set_title(f"Distribución de '{column}'")

plt.suptitle('Distribución de categorías de las v. categóricas')
plt.tight_layout()
plt.show()
```

## 2.3. Relaciones entre variables

```python
_, ax = plt.subplots(2, 1, figsize=(16, 9))

ax[0].scatter(df['recclass'], df['mass (g)'], alpha=0.75)
ax[0].set_title("Relación entre 'recclass' y 'mass (g)'")

ax[1].scatter(df['year'], df['mass (g)'], alpha=0.75)
ax[1].set_title("Relación entre 'year' y 'mass (g)'")

plt.show()
```
